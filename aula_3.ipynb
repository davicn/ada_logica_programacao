```python
numero_secreto = 42

contador = 0

while contador < 10:
  tentativa = int(input('Adivinhe o número secreto: '))
  contador += 1
  if tentativa != numero_secreto:
    print('Errou!')
    continue
  print('Acertou!')
  break
else:
  print('Acabaram as tentativas. Você perdeu.')
```
